```python
import json
from collections import Counter
from pprint import pprint


FILES = [
    r"E:\Projects\CheckThat 2026\task 2\data\train.json",
    r"E:\Projects\CheckThat 2026\task 2\data\validation.json",
    r"E:\Projects\CheckThat 2026\task 2\data\test.json"
]


def inspect_json(file_path):
    print("\n" + "=" * 100)
    print(f"FILE: {file_path}")
    print("=" * 100)

    with open(file_path, "r", encoding="utf-8") as f:
        data = json.load(f)

    print("\nTop-level type:")
    print(type(data))

    # If JSON is a list
    if isinstance(data, list):
        print(f"\nNumber of records: {len(data)}")

        if len(data) > 0:
            first = data[0]

            print("\nFirst record type:")
            print(type(first))

            if isinstance(first, dict):
                print("\nKeys in first record:")
                print(list(first.keys()))

                # print("\nSample record:")
                # pprint(first)

                # Try to find labels
                possible_label_keys = [
                    "label",
                    "gold_label",
                    "verdict",
                    "classification",
                    "answer",
                    "claim_type"
                ]

                for key in possible_label_keys:
                    if key in first:
                        labels = [x.get(key) for x in data if key in x]

                        print(f"\nDistribution of '{key}':")
                        pprint(Counter(labels))

    # If JSON is a dictionary
    elif isinstance(data, dict):
        print("\nTop-level keys:")
        print(list(data.keys()))

        print("\nFirst few key-value pairs:")

        for i, (k, v) in enumerate(data.items()):
            print(f"\n{k}:")
            pprint(v)

            if i == 4:
                break

    print("\nInspection completed.")


for file in FILES:
    inspect_json(file)
```
